# The **os** module

The OS module in Python provides a way of using operating system dependent functionality. 

The functions that the OS module provides allows you to interface with the underlying operating system that Python is running on – be that Windows, Mac or Linux. 

You can find important information about your location or about the process.

```
import os
```

---
Get the users environment

```
os.environ
```

---
Return information identifying the current operating system.

```
os.uname()
```

---
Returns the real process ID of the current process.

```
os.getpid()
```

---
Return the current process’s user id.

```
os.getuid()
```

---
Returns the current working directory.

```
os.getcwd()
```

---
Create a directory

```
os.mkdir("my_dir")
```

---
Rename the file or directory.

```
os.rename("my_dir", "my_new_dir")
```

---
List of files and directories in the current working directory

```
os.listdir()
```

---
List of all files and directories in the specified directory.

```
os.listdir("my_new_dir")
```

---
Change the current working directory to another directory

```
os.chdir("my_new_dir")
os.getcwd()
```

---
Change the current working directory to parent directory

```
os.chdir("..")
os.getcwd()
```

---
Remove (delete) the file path.

```
os.remove("my_file.txt")
```

---
Remove (delete) the directory path.

```
os.rmdir("my_new_dir")
```
