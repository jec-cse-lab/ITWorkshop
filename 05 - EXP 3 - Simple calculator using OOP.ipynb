```
class Calculator:
  pass
```

**Test the Calculator class below** -

```
c = Calculator(10)
print(c.addition(5))  # This should print 15 (10 + 5)
print(c.multiply(3))  # This should print 45 (15 x 3)
print(c.substract(2)) # This should print 43 (45 - 2)

print("The current value is ",c.getVal()) # This should print 43
```
